##Create a machine learning model that can classify the species of an iris flower based on its sepal and petal length and width.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
# Load the Iris dataset
iris = load_iris()
```

```python
X = iris.data
y = iris.target
feature_names = iris.feature_names

# Split the data into training and testing sets
test_size = 0.2
random_state = 42

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)

# Print the shape of the training and testing sets
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
```

```python
# Scatter plots
def scatter_plot(x_index, y_index):
    plt.figure(figsize=(8, 6))
    for target, target_name in enumerate(iris.target_names):
        plt.scatter(X[y == target, x_index], X[y == target, y_index], label=target_name)
    plt.xlabel(feature_names[x_index])
    plt.ylabel(feature_names[y_index])
    plt.title(f"Scatter plot of {feature_names[x_index]} vs {feature_names[y_index]}")
    plt.legend()
    plt.show()

# Scatter plot of sepal length vs sepal width
scatter_plot(0, 1)

# Scatter plot of petal length vs petal width
scatter_plot(2, 3)
```

```python
# Histograms
def histogram_plot(x_index):
    plt.figure(figsize=(8, 6))
    for target, target_name in enumerate(iris.target_names):
        plt.hist(X[y == target, x_index], bins=20, alpha=0.5, label=target_name)
    plt.xlabel(feature_names[x_index])
    plt.ylabel("Frequency")
    plt.title(f"Histogram of {feature_names[x_index]}")
    plt.legend()
    plt.show()

# Histograms of all features
for i in range(len(feature_names)):
    histogram_plot(i)
```

```python
#Training the dataset using KNN algorithm
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

# Make predictions on the testing data
y_pred = knn.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Generate classification report and confusion matrix
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
# New data to make predictions
new_data = np.array([[5.1, 3.5, 1.4, 0.2],   # Sample 1
                     [6.5, 3.0, 5.2, 2.0],   # Sample 2
                     [7.2, 3.6, 6.1, 2.5]])  # Sample 3

# Make predictions on new data
new_predictions = knn.predict(new_data)

# Convert numeric predictions to class names
class_names = iris.target_names
predicted_classes = [class_names[prediction] for prediction in new_predictions]

# Print the predictions
for i, sample in enumerate(new_data):
    print(f"Sample {i+1}: Predicted class - {predicted_classes[i]}")
```
